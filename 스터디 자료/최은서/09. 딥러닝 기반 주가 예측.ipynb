## 퍼셉트론 알고리즘 회로

```python
##1. AND
def AND(x1, x2):
    w1=0.5
    w2=0.5
    theta=0.7
    if w1*x1+w2*x2>theta:
        return 1
    else:
        return 0
print(AND(0, 0))
print(AND(0, 1))
print(AND(1, 0))
print(AND(1, 1))
```

```python
##2. NAND
def NAND(x1, x2):
    w1=-0.5
    w2=-0.5
    theta=-0.7
    if w1*x1+w2*x2>theta:
        return 1
    else:
        return 0
print(NAND(0, 0))
print(NAND(0, 1))
print(NAND(1, 0))
print(NAND(1, 1))
```

```python
##3. OR
def OR(x1, x2):
    w1=0.5
    w2=0.5
    theta=0.2
    if w1*x1+w2*x2>theta:
        return 1
    else:
        return 0
print(OR(0, 0))
print(OR(0, 1))
print(OR(1, 0))
print(OR(1, 1))
```

## 활성화 함수

```python
##계단함수 0 또는 1 출력
import numpy as np
import matplotlib.pyplot as plt

def stepfunc(x):
    return np.where(x <= 0, 0, 1)##0보다 작거나 같으면 0을 반환 그렇지 않으면 1


x = np.arange(-10, 10, 0.1)
y = stepfunc(x)

plt.plot(x, y)
plt.title('step function')
plt.show()
##XXe+b 는 XX^b 를 의미한다.
```

```python
print(x)
##XXe+b 는 XX^b 를 의미한다.
print(y)
```

```python
##시그모이드 함수 0~1사이의 연속적인 실수 출력(복잡한 신경망 계산 가능) 1/1+e^(-x)
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x)) 

x = np.arange(-10, 10, 0.1)
y = sigmoid(x) 

plt.plot(x, y)
plt.title('sigmoid function')
plt.show()
```

```python
print(x)
##XXe+b 는 XX^b 를 의미한다.
print(y)
```

```python
##tanh 함수 -1~1 사이의 값을 출력 e^x-e^(-x)/e^x+e^(-x)
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))  

x = np.arange(-10, 10, 0.1) 
y = tanh(x) 

plt.plot(x, y)
plt.title('tanh function')
plt.show()
```

```python
print(x)
##XXe+b 는 XX^b 를 의미한다.
print(y)
```

```python
##ReLu 함수 0~무한대까지 출력
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)##0이하이면 0을 그렇지 않으면 그대로 출력

x = np.arange(-10, 10, 0.1)
y = relu(x)

plt.plot(x, y)
plt.title('ReLU function')
plt.show()
```

```python
print(x)
##XXe+b 는 XX^b 를 의미한다.
print(y)
```

```python
##소프트맥스 함수 입력받은 값들을 0~1 사이 값으로 정규화. 출력값들의 총합은 1이 된다.
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
```

## 다층 퍼셉트론

```python
##은닉층 계산
import numpy as np
X = np.array([10, 20])  # ①
W1 = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
B1 = np.array([1, 2, 3])  # ③

def sigmoid(x): 
    return 1 / (1 + np.exp(-x))

A1 = np.dot(X, W1) + B1 ##dot은 matrix 내적연산,,
Z1 = sigmoid(A1) ##활성화 함수 : signoid

print('A1 :', A1)
print('Z1 :', Z1)
```

```python
##출력층 계산

W2 = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
B2 = np.array([0.1, 0.2])

A2 = np.dot(Z1, W2) + B2##출력층의 입력값(은닉층:Z1 * 가중치:W2 + 편향:B2)
Y = sigmoid(A2) ##활성화 함수 : signoid

print('A2 :', A2)
print('Y  :', Y)
```

## 선형 회귀 문제

```python
import matplotlib.pylab as plt
import tensorflow as tf

##h(x)=wx+b
##y=x+1
##W=0.7, b=0.7로 초기화(초기값과 실제 wb가 작을수록 학습에 걸리는 시간 줄어듬.)
x_data = [1, 2, 3, 4, 5]
y_data = [2, 3, 4, 5, 6]

w = tf.Variable(0.7)
b = tf.Variable(0.7)
learn_rate = 0.01##학습률 : 0.01~0.001사이. 너무 크면 오버슈팅(비용이 커짐), 너무 작으면 시간이 오래 걸림.
########################################################################################################
##비용, 손실=예측값-실젯값 제곱의 평균(=오차제곱평균)

print(f'step|    w|    b| cost')
print(f'----|-----|-----|-----')

for i in range(1, 1101): ##1100회의 학습 수행.
    with tf.GradientTape() as tape:
        hypothesis = w * x_data + b
        cost = tf.reduce_mean((hypothesis - y_data)**2) ##손실비용, tf.losses.mean_squared_error(y, y_hat)도 동일
    # a = a - b##미분해서 dw, db =>최적의 d, b값 찾기
    dw, db = tape.gradient(cost, [w, b])
    
    #학습
    w.assign_sub(learn_rate * dw) 
    b.assign_sub(learn_rate * db) 
    
    if i in [1, 3, 5, 10, 1000, 1100]:
        print(f"{i:4d}| {w.numpy():.2f}| {b.numpy():.2f}| {cost:.2f}")
        plt.figure(figsize=(7, 7))
        plt.title(f'[Step {i:d}]  h(x) = { w.numpy():.2f}x + {b.numpy():.2f}')
        plt.plot(x_data, y_data, 'o') # ⑥
        plt.plot(x_data, w * x_data + b, 'r', label='hypothesis') # ⑦
        plt.xlabel('x_data')
        plt.ylabel('y_data')
        plt.xlim(0, 6)
        plt.ylim(1, 7)
        plt.legend(loc='best')
        plt.show()
        
##w, b값이 실젯값인 1에 도달
```
